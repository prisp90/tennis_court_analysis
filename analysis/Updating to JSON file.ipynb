```python
import pandas as pd
import json
import openpyxl
import math

json_path = '/Users/pumpedup_py/Downloads/CSE6242-main/data/TEST.json'
df = pd.read_json(json_path)

xl_path = "/Users/pumpedup_py/Downloads/TennisCourts.xlsx"
wb = openpyxl.load_workbook(xl_path)
main_sheet = wb["summary"]
```

```python
# store every value from 'summary' sheet from 'TennisCourts.xlsx' file.

city=[]
total_num_courts=[]
population=[]
tennis_population=[]
match_est=[]
min_match=[]
max_match=[]
min_demanded_courts_s=[]
min_demanded_courts_d=[]
max_demanded_courts_s=[]
max_demanded_courts_d=[]
sd_min_s=[]
sd_min_d=[]
sd_max_s=[]
sd_max_d=[]

for i in range(2, main_sheet.max_row+1):
    city_value = main_sheet.cell(row=i, column=1).value
    total_num_courts_value = main_sheet.cell(row=i, column=2).value
    population_value = main_sheet.cell(row=i, column=4).value
    tennis_population_value = main_sheet.cell(row=i, column=5).value
    match_est_value = main_sheet.cell(row=i, column=6).value
    min_match_value = main_sheet.cell(row=i, column=7).value
    max_match_value = main_sheet.cell(row=i, column=8).value
    min_demanded_courts_s_value = main_sheet.cell(row=i, column=9).value
    min_demanded_courts_d_value = main_sheet.cell(row=i, column=10).value
    max_demanded_courts_s_value = main_sheet.cell(row=i, column=11).value
    max_demanded_courts_d_value = main_sheet.cell(row=i, column=12).value
    sd_min_s_value = main_sheet.cell(row=i, column=13).value
    sd_min_d_value = main_sheet.cell(row=i, column=14).value
    sd_max_s_value = main_sheet.cell(row=i, column=15).value
    sd_max_d_value = main_sheet.cell(row=i, column=16).value
    
    try:
        tennis_population_value = math.ceil(tennis_population_value)
        match_est_value = math.ceil(match_est_value)
        min_match_value = math.ceil(min_match_value)
        max_match_value = math.ceil(max_match_value)
        min_demanded_courts_s_value = math.ceil(min_demanded_courts_s_value)
        min_demanded_courts_d_value = math.ceil(min_demanded_courts_d_value)
        max_demanded_courts_s_value = math.ceil(max_demanded_courts_s_value)
        max_demanded_courts_d_value = math.ceil(max_demanded_courts_d_value)
        sd_min_s_value = math.ceil(sd_min_s_value * (10 ** 4)) / (10 ** 4)
        sd_min_d_value = math.ceil(sd_min_d_value * (10 ** 4)) / (10 ** 4)
        sd_max_s_value = math.ceil(sd_max_s_value * (10 ** 4)) / (10 ** 4)
        sd_max_d_value = math.ceil(sd_max_d_value * (10 ** 4)) / (10 ** 4)
    except:
        pass
    
    city.append(city_value)
    total_num_courts.append(total_num_courts_value)
    population.append(population_value)
    tennis_population.append(tennis_population_value)
    match_est.append(match_est_value)
    min_match.append(min_match_value)
    max_match.append(max_match_value)
    min_demanded_courts_s.append(min_demanded_courts_s_value)
    min_demanded_courts_d.append(min_demanded_courts_d_value)
    max_demanded_courts_s.append(max_demanded_courts_s_value)
    max_demanded_courts_d.append(max_demanded_courts_d_value)
    sd_min_s.append(sd_min_s_value)
    sd_min_d.append(sd_min_d_value)
    sd_max_s.append(sd_max_s_value)
    sd_max_d.append(sd_max_d_value)
```

```python
# check if there's a missing data across the categories.

print(len(city))
print(len(total_num_courts))
print(len(population))
print(len(tennis_population))
print(len(match_est))
print(len(min_match))
print(len(max_match))
print(len(min_demanded_courts_s))
print(len(min_demanded_courts_d))
print(len(max_demanded_courts_s))
print(len(max_demanded_courts_d))
print(len(sd_min_s))
print(len(sd_min_d))
print(len(sd_max_s))
print(len(sd_max_d))
```

```python
# creates a new Pandas DataFrame by appending data from multiple lists
# and remove unnecessary rows.

df_new = pd.DataFrame()

for i in range(len(city)):
    new_data = {'CITY': city[i],
                'TOTAL_NUM_COURTS': total_num_courts[i],
                'POPULATION': population[i],
                'TENNIS_POPULATION': tennis_population[i],
                'MATCH_EST': match_est[i],
                'MIN_MATCH': min_match[i],
                'MAX_MATCH': max_match[i],
                'MIN_DEMANDED_COURS_S': min_demanded_courts_s[i],
                'MIN_DEMANDED_COURS_D': min_demanded_courts_d[i],
                'MAX_DEMANDED_COURS_S': max_demanded_courts_s[i],
                'MAX_DEMANDED_COURS_D': max_demanded_courts_d[i],
                'SD_MIN_S': sd_min_s[i],
                'SD_MIN_D': sd_min_d[i],
                'SD_MAX_S': sd_max_s[i],
                'SD_MAX_D': sd_max_d[i]
               }
    df_new = df_new.append(new_data, ignore_index=True)

df_new = df_new.drop(df_new.index[-1])
df_new = df_new.drop(df_new.index[-1])
df_new = df_new.drop(df_new.index[-1])
```

```python
# Read the existing JSON file into a DataFrame
right_index = []

for i in range(0, 215):
    for j in range(0, 218):
        if df['features'][i]['properties']['NAME'] == df_new['CITY'][j]:
            right_index.append(i)
        else:
            pass
```

```python
# excluding city if there's not in the dataframe.

right_place_index = list(range(0, 215))
right_place_index = [i for i in right_place_index if i not in right_index]

for i in right_place_index:
    print(df['features'][i]['properties']['NAME'])
```

```python
# updates the properties of features in a GeoJSON file with the data from a DataFrame, 
# based on matching city names between the two.

for i in range(0, 215):
    for j in range(0, 218):
        if df['features'][i]['properties']['NAME'] == df_new['CITY'][j]:
            new_data = {
                'TOTAL_NUM_COURTS': total_num_courts[i],
                'POPULATION': population[i],
                'TENNIS_POPULATION': tennis_population[i],
                'MATCH_EST': match_est[i],
                'MIN_MATCH': min_match[i],
                'MAX_MATCH': max_match[i],
                'MIN_DEMANDED_COURS_S': min_demanded_courts_s[i],
                'MIN_DEMANDED_COURS_D': min_demanded_courts_d[i],
                'MAX_DEMANDED_COURS_S': max_demanded_courts_s[i],
                'MAX_DEMANDED_COURS_D': max_demanded_courts_d[i],
                'SD_MIN_S': sd_min_s[i],
                'SD_MIN_D': sd_min_d[i],
                'SD_MAX_S': sd_max_s[i],
                'SD_MAX_D': sd_max_d[i]
               }
            df['features'][i]['properties'].update(new_data)
            
        else:
            pass

df.to_json(json_path)
```
